```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data=pd.read_csv('/home/abhishek/Downloads/data.csv')
data.head()
data['id'] = data['id'].astype('category').cat.codes
data['diagnosis'] = data['diagnosis'].astype('category').cat.codes
data.head()
```

```python
plt.scatter(data['id'],data['radius_mean'])
plt.show()
sns.distplot(data['radius_mean'])
sns.plt.show()
data['radius_mean'].hist()
plt.show()
```
